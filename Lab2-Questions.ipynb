#### Practice Exercise 2
Read the coordinates of the capital cities of the world from the country-capital-lat-long-population.csv file. Use NumPy to print the name and coordinates of the capital city closest to the Saudi Arabia capital - Riyadh.

Note that:

1. The Country Name for Saudi Arabia is given as Saudi Arabia in the data.
2. The ‘closeness’ of capital cities from the Saudi Arabia capital is based on the Euclidean distance of their coordinates to those of the Saudi Arabia capital.


Hints:

1. Use read_html() from the Pandas library to read the table.
2. Use the to_numpy() function of the Pandas DataFrame class to convert a DataFrame to a Numpy array
3. Use broadcasting to compute the euclidean distance of capital cities from Riyadh.

```python
import pandas as pd
import numpy as np
country_capital = pd.read_csv("country-capital-lat-long-population.csv")

riyadh_data = country_capital[country_capital['Country'] == 'Saudi Arabia']
riyadh_lat = riyadh_data['Latitude'].values[0]
riyadh_long = riyadh_data['Longitude'].values[0]
capitals_lat = country_capital['Latitude'].to_numpy()
capitals_long = country_capital['Longitude'].to_numpy()
distances = np.sqrt((capitals_lat - riyadh_lat) ** 2 + (capitals_long - riyadh_long) ** 2)
closest_index = np.argmin(distances)
closest_capital = country_capital['Capital City'].values[closest_index]
closest_lat = capitals_lat[closest_index]
closest_long = capitals_long[closest_index]
print(f"The closest capital city to Riyadh is {closest_capital} at coordinates ({closest_lat}, {closest_long}).")

```

#### Practical Exercise 2

Use NumPy to:

1. Print the names of the countries of the top 10 capital cities closest to the US capital - Washington DC.

2. Create and print a NumPy array containing the coordinates of the top 10 cities.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("country-capital-lat-long-population.csv")
df.columns = df.columns.str.strip()
dc_lat = df[df['Country'] == 'United States of America']['Latitude'].values[0]
dc_long = df[df['Country'] == 'United States of America']['Longitude'].values[0]

capitals_lat = df['Latitude'].to_numpy()  
capitals_long = df['Longitude'].to_numpy() 
countries = df['Country'].to_numpy() 

distances = np.sqrt((capitals_lat - dc_lat) ** 2 + (capitals_long - dc_long) ** 2)

closest_indices = np.argsort(distances)[:10]

print("Top 10 closest capital cities to Washington D.C.:")
for index in closest_indices:
    capital_name = df['Capital City'].values[index]  
    print(f"Country: {countries[index]}, Capital: {capital_name}")

top_10_coordinates = np.array(list(zip(capitals_lat[closest_indices], capitals_long[closest_indices])))
print("\nCoordinates of the top 10 closest capital cities (Latitude, Longitude):")
print(top_10_coordinates)

```
